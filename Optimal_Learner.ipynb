## Transition matrices

```python
def plot_slices(grid, joint, cmap, step=2, var=""):
    pal = sns.blend_palette(sns.mpl_palette(cmap, 6), as_cmap=True)
    f, axes = plt.subplots(4, 4, figsize=(10, 10), sharex=True, sharey=True)
    xx, yy = np.meshgrid(grid, grid)
    for k, ax in zip(np.arange(learner._k_size) * step, axes.flat):
        ax.contour(xx, yy, joint[:, :, k], cmap=pal, vmin=joint.min(), vmax=joint.max())
    if var:
        for ax in axes[-1, :]:
            ax.set_xlabel(r"$%s_{i+1}$" % var, size=16)
        for ax in axes[:, 0]:
            ax.set_ylabel(r"$%s_{i}$" % var, size=16)
    plt.tight_layout()
```

```python
plot_slices(learner.p_grid, learner._p_trans, "GnBu_d", var="p")
```

```python
plot_slices(learner.I_grid, learner._I_trans, "OrRd_d", var="\log\/\ v^{-1}")
```

## Example model fits

```python
def residual_plot(learner, ground_truth):
    f, ax = plt.subplots(figsize=(9, 2))
    ax.plot(learner.p_hats - ground_truth, c="#444444", lw=1.5)
    ax.axhline(0, ls="--", lw=1.5, c="#777777")
    ax.set(xlabel="Trial", ylabel="Residual")
```

### Static probability

```python
static_p = np.ones(500) * .65
static_y = stats.binom.rvs(1, static_p)
```

```python
learner.reset()
learner.fit(static_y)
```

```python
learner.plot_history(ground_truth=static_p)
```

```python
residual_plot(learner, static_p);
```

### Nonstationary probabilities, static volatility

```python
moving_p = np.repeat([.75, .25, .75, .25, .75, .25], 100)
moving_y= stats.binom.rvs(1, moving_p)
```

```python
learner.reset()
learner.fit(moving_y)
```

```python
learner.plot_history(ground_truth=moving_p)
```

```python
residual_plot(learner, moving_p);
```

### Nonstationary probabilities and volatility

```python
metavol_p = np.repeat([.75, .75, .75, .25, .75, .25], 100)
metavol_y= stats.binom.rvs(1, metavol_p)
```

```python
learner.reset()
learner.fit(metavol_y)
```

### Sinusoidal probabilities

```python
x = np.linspace(0, 60, 600)
sin_p = (np.sin(x / 3) + 1.5) / 3
sin_y = stats.binom.rvs(1, sin_p)
```

```python
learner.reset()
learner.fit(sin_y)
```

```python
learner.plot_history(ground_truth=sin_p)
```

```python
residual_plot(learner, sin_p)
```
