# Bayesian probability learner

This notebook uses a pure Python implementation of the computational model in  [this paper](http://www.ncbi.nlm.nih.gov/pubmed/17676057).

```python
from __future__ import division
from IPython.display import display, clear_output
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot, contourf
import seaborn as sns
```

```python
import optlearner
```

### Define the data

This is a basic schedule and (reproducibly) random data to fit the model with

```python
sched = [.75] * 60 + [.2] * 20 + [.8] * 20 + [.2] * 20
binom = np.random.RandomState(7).binomial
reward = np.concatenate([binom(1, .75, 60), binom(1, .2, 20), binom(1, .8, 20), binom(1, .2, 20)])
```

### Initialize and fit the learner

```python
reload(optlearner)
```

```python
learner = optlearner.OptimalLearner()
```

```python
learner.fit(reward)
```

```python
p_grid = learner.p_grid
v_grid = learner.v_grid
```

### Visualize the model

Plot the posterior means for both $p$ and $v$ over time

```python
learner.plot_history(ground_truth=sched)
```

Plot the final joint distribution $P(~p, v~|~y_{\le{i}}~)$

```python
learner.plot_joint()
```

Animate the posterior densities on $p$ and $v$ as it evolves over time

```python
red, blue, green = sns.husl_palette(3)
```

```python
ax_p = plt.subplot(211, ylim=(0, .1))
ax_p.set_ylabel("$P\,(p)$", size=16)
ax_v = plt.subplot(212, xlim=(v_grid.min(), v_grid.max()), ylim=(0, .06))
ax_v.set_ylabel("$P\,(v)$", size=16)
for i, (p_d, v_d) in enumerate(zip(learner.p_dists, learner.v_dists)):
    p_line, = ax_p.plot(p_grid, p_d, c=blue)
    v_line, = ax_v.plot(v_grid, v_d, c=green)
    ax_p.set_title("Trial %d" % i)
    display(plt.gcf())
    clear_output()
    if i < len(sched) - 1:
        p_line.remove()
        v_line.remove()
```
